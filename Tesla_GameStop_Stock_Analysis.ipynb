```python
import yfinance as yf
import pandas as pd
import requests
from bs4 import BeautifulSoup
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import warnings
warnings.filterwarnings('ignore')
```

```python
# Extract Tesla stock data using yfinance
tesla_stock = yf.Ticker("TSLA")

# Get historical data for Tesla
tesla_data = tesla_stock.history(period="max")

# Display the first few rows
print("Tesla Stock Data:")
print(f"Data shape: {tesla_data.shape}")
print("\nFirst 5 rows:")
tesla_data.head()
```

```python
# Reset index to make Date a column
tesla_data.reset_index(inplace=True)
print("Tesla data after resetting index:")
tesla_data.head()
```

```python
# Extract Tesla revenue data using web scraping
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Add headers to avoid being blocked
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
html_data = requests.get(url, headers=headers).text
soup = BeautifulSoup(html_data, "html.parser")

# Find all tables on the page
tables = soup.find_all("table")
tesla_revenue = None

for table in tables:
    table_text = table.get_text()
    if any(keyword in table_text for keyword in ["Tesla Quarterly Revenue", "Tesla Annual Revenue", "Quarterly Revenue", "Annual Revenue", "Revenue"]):
        tesla_revenue = pd.read_html(str(table))[0]
        break

# Clean the revenue data
if tesla_revenue is not None:
    tesla_revenue = tesla_revenue.iloc[:, :2]  # Take first two columns
    tesla_revenue.columns = ["Date", "Revenue"]
    tesla_revenue["Revenue"] = tesla_revenue["Revenue"].astype(str).str.replace("$", "").str.replace(",", "")
    tesla_revenue.dropna(inplace=True)
    tesla_revenue = tesla_revenue[tesla_revenue["Revenue"] != ""]
    tesla_revenue["Revenue"] = pd.to_numeric(tesla_revenue["Revenue"], errors='coerce')
    tesla_revenue.dropna(inplace=True)
    print("Tesla Revenue Data:")
    print(f"Data shape: {tesla_revenue.shape}")
    tesla_revenue.head()
else:
    print("❌ Tesla revenue data not found.")
    # Backup data in case scraping fails
    tesla_revenue = pd.DataFrame({
        'Date': ['2023-Q4', '2023-Q3', '2023-Q2', '2023-Q1', '2022-Q4'],
        'Revenue': [25167, 23350, 24927, 23329, 24318]
    })
    print("Using backup Tesla revenue data:")
    tesla_revenue.head()
```

```python
# Extract GameStop stock data using yfinance
gamestop_stock = yf.Ticker("GME")

# Get historical data for GameStop
gamestop_data = gamestop_stock.history(period="max")

# Display the first few rows
print("GameStop Stock Data:")
print(f"Data shape: {gamestop_data.shape}")
print("\nFirst 5 rows:")
gamestop_data.head()
```

```python
# Reset index to make Date a column
gamestop_data.reset_index(inplace=True)
print("GameStop data after resetting index:")
gamestop_data.head()
```

```python
# Extract GameStop revenue data using web scraping
url = "https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue"
html_data = requests.get(url, headers=headers).text
soup = BeautifulSoup(html_data, "html.parser")

# Find all tables on the page
tables = soup.find_all("table")
gamestop_revenue = None

for table in tables:
    table_text = table.get_text()
    if any(keyword in table_text for keyword in ["GameStop Quarterly Revenue", "GameStop Annual Revenue", "Quarterly Revenue", "Annual Revenue", "Revenue"]):
        gamestop_revenue = pd.read_html(str(table))[0]
        break

# Clean the revenue data
if gamestop_revenue is not None:
    gamestop_revenue = gamestop_revenue.iloc[:, :2]  # Take first two columns
    gamestop_revenue.columns = ["Date", "Revenue"]
    gamestop_revenue["Revenue"] = gamestop_revenue["Revenue"].astype(str).str.replace("$", "").str.replace(",", "")
    gamestop_revenue.dropna(inplace=True)
    gamestop_revenue = gamestop_revenue[gamestop_revenue["Revenue"] != ""]
    gamestop_revenue["Revenue"] = pd.to_numeric(gamestop_revenue["Revenue"], errors='coerce')
    gamestop_revenue.dropna(inplace=True)
    print("GameStop Revenue Data:")
    print(f"Data shape: {gamestop_revenue.shape}")
    gamestop_revenue.head()
else:
    print("❌ GameStop revenue data not found.")
    # Backup data in case scraping fails
    gamestop_revenue = pd.DataFrame({
        'Date': ['2023-Q4', '2023-Q3', '2023-Q2', '2023-Q1', '2022-Q4'],
        'Revenue': [2282, 1164, 1237, 1378, 1299]
    })
    print("Using backup GameStop revenue data:")
    gamestop_revenue.head()
```

```python
def make_graph(stock_data, revenue_data, stock_name):
    """
    Create a dashboard with two subplots:
    1. Stock price over time
    2. Revenue over time
    """
    fig = make_subplots(rows=2, cols=1,
                        shared_xaxes=True,
                        subplot_titles=(f"{stock_name} Stock Price", f"{stock_name} Revenue"),
                        vertical_spacing=0.1)

    # Add stock price trace
    fig.add_trace(
        go.Scatter(x=stock_data['Date'],
                  y=stock_data['Close'],
                  mode='lines',
                  name=f"{stock_name} Stock Price",
                  line=dict(color='blue')),
        row=1, col=1
    )

    # Add revenue trace
    fig.add_trace(
        go.Scatter(x=revenue_data['Date'],
                  y=revenue_data['Revenue'],
                  mode='lines+markers',
                  name=f"{stock_name} Revenue",
                  line=dict(color='red')),
        row=2, col=1
    )

    # Update layout
    fig.update_layout(
        title=f"{stock_name} Stock Price vs Revenue Dashboard",
        height=800,
        showlegend=True,
        template='plotly_white'
    )

    # Update y-axes labels
    fig.update_yaxes(title_text="Stock Price ($)", row=1, col=1)
    fig.update_yaxes(title_text="Revenue ($ Millions)", row=2, col=1)
    fig.update_xaxes(title_text="Date", row=2, col=1)

    return fig
```

```python
# Create Tesla dashboard
tesla_fig = make_graph(tesla_data, tesla_revenue, "Tesla")
tesla_fig.show()

print("Tesla Dashboard created successfully!")
print(f"Tesla stock data points: {len(tesla_data)}")
print(f"Tesla revenue data points: {len(tesla_revenue)}")
```

```python
# Create GameStop dashboard
gamestop_fig = make_graph(gamestop_data, gamestop_revenue, "GameStop")
gamestop_fig.show()

print("GameStop Dashboard created successfully!")
print(f"GameStop stock data points: {len(gamestop_data)}")
print(f"GameStop revenue data points: {len(gamestop_revenue)}")
```
